```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Access the features and target variables
X = iris.data   # Features (sepal length, sepal width, petal length, petal width)
y = iris.target # Target variable (species: 0 = setosa, 1 = versicolor, 2 = virginica)

# Print the shape of the dataset
print("Dataset shape:", X.shape)
```

```python
import pandas as pd
df = pd.DataFrame({"sl": X[:,0],
                  "sw" : X[:,1],
                  "pl" : X[:,2],
                  "pw" : X[:,3]})

import seaborn as sns
sns.pairplot(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.pl, df.pw)
plt.title('Petal width as a function of petal length')
plt.xlabel('pl (petal length)')
plt.ylabel('pw (petal width)')
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=9)

plt.scatter(train.pl, train.pw, 3, 'b', label = 'train')
plt.scatter(test.pl, test.pw, 3, 'r', label = 'test')

plt.title('Petal width as a function of petal length')
plt.xlabel('pl (petal length)')
plt.ylabel('pw (petal width)')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression

# Create and fit the linear regression model
model = LinearRegression()

# model.fit(train.pl, train.pw)
"Expected 2D array, got 1D array instead:\narray={}.\n"
"Reshape your data either using array.reshape(-1, 1)"

import numpy as np
X_train = np.array(train.pl).reshape(-1, 1)
y_train = np.array(train.pw).reshape(-1, 1)

X_test = np.array(test.pl).reshape(-1, 1)
y_test = np.array(test.pw).reshape(-1, 1)

# model = LinearRegression()
model.fit(X_train, y_train)
y_pred_visu = model.predict([[0],[7]])

y_pred = model.predict(X_test)
```

```python
plt.scatter(train.pl, train.pw, 3, 'b', label = 'train')
plt.plot([0,7], y_pred_visu, 'k--', label = 'fit', linewidth = 0.5)
plt.scatter(test.pl, test.pw, 3, 'r', label = 'test')
plt.scatter(test.pl, y_pred, 15, 'g', marker = 'x', label = 'predict')

plt.title('Petal width as a function of petal length')
plt.xlabel('pl (petal length)')
plt.ylabel('pw (petal width)')
plt.legend()
```

```python
sum((y_pred - y_test)**2)/len(y_test)  # MSE
```
